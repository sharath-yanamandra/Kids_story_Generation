```python
def get_story(url):
  """It gets url and get title and story on the page.
  """
  response = requests.get(url)
  soup = BeautifulSoup(response.text, "html.parser")
  st = soup.find(
      name = 'h1'
  ).get_text()

  full_story = ""
  story_text = soup.find(name="div", class_="entry-content")
  for p_tag in story_text.findAll(name='p'):
      full_story += p_tag.get_text()
  
  return (st, full_story)
```

```python
def get_story_url_from_page(url):
  url_list = []
  response = requests.get(url)
  soup = BeautifulSoup(response.text, "html.parser")
  main_story_box = soup.findAll(name='a', class_='margin20 inline-block')
  for story_0 in main_story_box:
    url_list.append(story_0.get('href'))
  
  return url_list
```

```python
# adding first page link
all_url_list = get_story_url_from_page("https://www.studentuk.com/category/bedtime-stories/")

# getting all the other links
for page in range(2, 26):
  all_url_list.extend(get_story_url_from_page(f"https://www.studentuk.com/category/bedtime-stories/page/{page}/"))
```

```python
title = []
story = []

for url in all_url_list:
  st, full_story = get_story(url)
  title.append(st)
  story.append(full_story)
```

```python
storiess["title"] = title
storiess["story"] = story
storiess["url"] = all_url_list
```
